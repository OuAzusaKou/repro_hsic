```python
import numpy as np
delta=5
def gausin_distance(x,y,delta=delta):
    H= torch.norm(x-y).cuda()
    distance = torch.exp(-H/2/(delta**2)).cuda()
    
    return distance
def gram_matrix(data,kernel=gausin_distance):
    #num=data.shape[0]
    #matrix=torch.zeros((num,num)).cuda()
    #for i in range (0,num):
    #   for j in range (0,num):
    #        matrix[i][j]=kernel(data[i],data[j])
    num=data.shape[0]
    datav3=torch.mm(data,torch.transpose(data, 0, 1).cuda()).cuda()
    datav1=torch.diag(datav3,0).cuda()
    #print(datav1.shape)
    #matrix=torch.sqrt(datav1+datav2-2*datav3)
    buf1=((-2)*datav3+datav1).cuda()
    buf2=torch.transpose(buf1, 0, 1).cuda()
    buf3=(buf2+datav1).cuda()
    #print(buf3)
    matrix = torch.exp(-buf3/2/(delta**2)).cuda()
    return matrix
def emerinal_hsic(X,Y):
    num=X.shape[0]
    #print(num)
    
    
    
    Kx=gram_matrix(X)
    
    Ky=gram_matrix(Y)
    
    H = torch.eye(num).cuda() - torch.ones((num,num), dtype = torch.float32).cuda() / num
    #print(H)
    hsic=1/(num-1)*torch.trace(torch.mm(torch.mm(torch.mm(Kx,H).cuda(),Ky).cuda(),H).cuda()).cuda()
    
    return hsic
import torch
import torchvision
from torchvision import transforms
import torch.nn.functional as F
import torchvision.datasets
from PIL import Image
batch_size=256
D_in=784
class MyMNIST(torchvision.datasets.MNIST):
    def __init__(self, root, train=True, transform=None, target_transform=None,download=False):
        
        super(MyMNIST,self).__init__(root, train, transform, target_transform,download)
        #print(self.data)
    def __getitem__(self, index):
        """
        Args:
            index (int): Index

        Returns:
            tuple: (image, target) where target is index of the target class.
        """
        img, target = self.data[index], int(self.targets[index])
        #print(target) 
        # doing this so that it is consistent with all other datasets
        # to return a PIL Image
        img = Image.fromarray(img.numpy(), mode='L')

        if self.transform is not None:
            img = self.transform(img)

        if self.target_transform is not None:
            target = self.target_transform(target)
            
        else:
            target=torch.as_tensor(target)
            target = F.one_hot(target,num_classes=10).cuda()

        return img, target
mnist_data=torchvision.datasets.MNIST(root='./MNIST/', train=True, transform=transforms.ToTensor(), target_transform=None, download=False)
data_loader = torch.utils.data.DataLoader(mnist_data,
                                          batch_size=batch_size,
                                          shuffle=True,
                                          num_workers=2)
D_in, H1,H2,H3,H4,H5,H6,H7, D_out = 784, 64,32,16,8,4,2,1,10
costs=[]
dtype=torch.float
continue_flag=1
PATH='./checkpoint_lamda1000'
w1 = torch.randn(D_in, H1, dtype=dtype, requires_grad=True,device="cuda")
w2 = torch.randn(H1, H2, dtype=dtype, requires_grad=True,device="cuda")
w3 = torch.randn(H2, H3, dtype=dtype, requires_grad=True,device="cuda")
w4 = torch.randn(H3, H4, dtype=dtype, requires_grad=True,device="cuda")
w5 = torch.randn(H4, H5, dtype=dtype, requires_grad=True,device="cuda")
w6 = torch.randn(H5, H6, dtype=dtype, requires_grad=True,device="cuda")
w7 = torch.randn(H6, H7, dtype=dtype, requires_grad=True,device="cuda")
w8 = torch.randn(H7, D_out, dtype=dtype, requires_grad=True,device="cuda")
if continue_flag:
    checkpoint = torch.load(PATH)
    w1=torch.tensor(checkpoint['w1'],dtype=dtype,requires_grad=True,device="cuda")
    w2=torch.tensor(checkpoint['w2'],dtype=dtype,requires_grad=True,device="cuda")
    w3=torch.tensor(checkpoint['w3'],dtype=dtype,requires_grad=True,device="cuda")
    w4=torch.tensor(checkpoint['w4'],dtype=dtype,requires_grad=True,device="cuda")
    w5=torch.tensor(checkpoint['w5'],dtype=dtype,requires_grad=True,device="cuda")
    w6=torch.tensor(checkpoint['w6'],dtype=dtype,requires_grad=True,device="cuda")
    w7=torch.tensor(checkpoint['w7'],dtype=dtype,requires_grad=True,device="cuda")
    w8=torch.tensor(checkpoint['w8'],dtype=dtype,requires_grad=True,device="cuda")
    
epoches=20
lamda=100
learning_rate=0.01
data_tsne_label=[]
data_tsne_rp1=[]
data_tsne_rp2=[]
for i in range(epoches): 
    for i_batch, (img,target) in enumerate(data_loader):
        img=img.reshape(-1,D_in).cuda()
        
        target=torch.as_tensor(target).cuda()
        num=target.shape[0]
        #print(num)
        target=target.reshape(num,1).cuda()
        #target = F.one_hot(target,num_classes=10).cuda()
        target=torch.zeros((num,10)).cuda().scatter_(1,target,1).cuda()
        target=torch.as_tensor(target,dtype=dtype).cuda()
        z1=torch.sigmoid(torch.mm(img,w1).cuda()).cuda()
        z2=torch.sigmoid(torch.mm(z1,w2).cuda()).cuda()
        z3=torch.sigmoid(torch.mm(z2,w3).cuda()).cuda()
        z4=torch.sigmoid(torch.mm(z3,w4).cuda()).cuda()
        z5=torch.sigmoid(torch.mm(z4,w5).cuda()).cuda()
        z6=torch.sigmoid(torch.mm(z5,w6).cuda()).cuda()
        z7=torch.sigmoid(torch.mm(z6,w7).cuda()).cuda()
        z8=torch.sigmoid(torch.mm(z7,w8).cuda()).cuda()
        for cnt in range(7):
            
            'cost'=emerinal_hsic('z'+str(i+1),img)-lamda*emerinal_hsic('z'+str(i+1),target)
        #print(cost1)
            cost1.backward(retain_graph=True)
        #print(w1.cuda().grad)
            with torch.no_grad():
                grad1=w1.grad
            #print('grad1',torch.norm(grad1))
            #w2 -= learning_rate * w2.grad

            # Manually zero the gradients after updating weights
                w1.grad.zero_()
            #w2.grad.zero_()
        cost2=emerinal_hsic(z2,img)-lamda*emerinal_hsic(z2,target)
        cost2.backward()
        with torch.no_grad():
            #w1 -= learning_rate * w1.grad
            #print('grad2',torch.norm(w2.grad))
            w2 -= learning_rate * w2.grad
            w1 -= learning_rate * grad1
        # Manually zero the gradients after updating weights
            w1.grad.zero_()
            w2.grad.zero_()
        if i_batch%50 ==0:
            data_tsne_rp2.append(z2.cpu().data.numpy())
            data_tsne_rp1.append(z1.cpu().data.numpy())
            data_tsne_label.append(target.argmax(dim=1, keepdim=True).cpu().data.numpy())
        print(i_batch)
        if i_batch %100 ==0:
            print('cost1',cost1,'\n cost2',cost2)
            torch.save({
            'w1': w1,
            'w2': w2,
            'data_tsne_label':data_tsne_label,
            'data_tsne_rp2':data_tsne_rp2,
            'data_tsne_rp1':data_tsne_rp1
                
            }, PATH)
            data_tsne_label=[]
            data_tsne_rp1=[]
            data_tsne_rp2=[]

            #for j,zz in enumerate(z2):
                #print(j,zz)
                #print(j,target[j])
            
        if i_batch % 500 == 0:
            costs.append(cost2)
plt.plot(costs)
plt.ylabel('cost2')
plt.xlabel('epochs (per 1)')
plt.title("Learning rate = " + str(learning_rate))
plt.show()        
    
```
